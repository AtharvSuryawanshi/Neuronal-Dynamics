```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
from scipy.integrate import odeint # type: ignore
import matplotlib.animation as animation 
from neuron_model import NeuronModel
```

```python
model = NeuronModel('saddle-node') # ['saddle-node', 'SNIC', 'subcritical_Hopf', 'supercritical_Hopf']
t = np.linspace(0, 10, 100)  # Time points
step_time = 2  # Step starts at t=2
step_duration = 3  # Lasts for 3 time units
baseline = 0  # Baseline current
amplitude = 1  # Amplitude of the step


dt = 0.005
T = 5
I_ext = 10*np.ones(int(T/dt)) #0.5*np.arange(0, T, dt)
a = model.simulate(T, dt, [-70, 0.3], I_ext)
# model.find_equlibrium_points(I_ext, [-70, 30], len(I_ext))
fig, axs = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]})
fig.suptitle("Hodgkin-Huxley model")
axs[0].plot(a[0], a[1][:, 0])
axs[0].set_xticks([])
axs[0].set_ylabel("Membrane potential (mV)")
axs[1].plot(a[0], I_ext)
axs[1].set_xlabel("Time (ms)")
axs[1].set_ylabel("I_ext ($uA/cm^2$)")
plt.tight_layout()
plt.show()
```

```python

# properties
equlibria = model.find_equlibrium_points(0, [-90, 20])
seperatrix = model.find_separatrix(0) # all other values are default
limit_cycle = model.find_limit_cycle(1)


plt.plot(a[1][:, 0], a[1][:, 1], alpha = 0.9)
plt.scatter(equlibria[0]['point'][0], equlibria[0]['point'][1], color='red')
plt.scatter(equlibria[1]['point'][0], equlibria[1]['point'][1], color='red')
plt.scatter(equlibria[2]['point'][0], equlibria[2]['point'][1], color='red')
# plt.plot(seperatrix[0], seperatrix[1], color='green', label='seperatrix')
plt.plot(limit_cycle[0], limit_cycle[1], color='blue', label='limit cycle')
plt.title("Phase plot") 
plt.xlabel("Membrane potential (mV)")   
plt.ylabel("n") 
plt.legend()
plt.show()
```

```python
T = 100
dt = 0.005  
X0 = [-70, 0.3]
I_ext = 0.5*np.arange(0, T, dt)
perts = [
    (20, 10, 0.0), # Voltage perturbation
    (80, 0, 0.5)   # n perturbation
]
plt.figure()
b = model.simulate_with_perturbations(T, dt, X0, I_ext, perts)
fig, axs = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]})
fig.suptitle("Hodgkin-Huxley model")
axs[0].plot(b[0], b[1][:, 0])
axs[0].set_xticks([])
axs[0].set_ylabel("Membrane potential (mV)")
axs[1].plot(b[0], I_ext)
axs[1].set_xlabel("Time (ms)")
axs[1].set_ylabel("I_ext ($uA/cm^2$)")
plt.tight_layout()
plt.show()

equlibria = model.find_equlibrium_points(0, [-90, 20])
    
plt.plot(b[1][:, 0], b[1][:, 1], alpha = 0.7)
plt.scatter(equlibria[0]['point'][0], equlibria[0]['point'][1], color='red')
plt.scatter(equlibria[1]['point'][0], equlibria[1]['point'][1], color='red')
plt.scatter(equlibria[2]['point'][0], equlibria[2]['point'][1], color='red')
plt.title("Phase plot") 
plt.xlabel("Membrane potential (mV)")   
plt.ylabel("n") 
plt.show()
```

```python
model = NeuronModel('subcritical_Hopf') # ['saddle-node', 'SNIC', 'subcritical_Hopf', 'supercritical_Hopf']
T = 100 
dt = 0.005
I_ext = 1*np.arange(0, T, dt)
a = model.simulate(T, dt, [-70, 0.3], I_ext)
fig, axs = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]})
fig.suptitle("Hodgkin-Huxley model")
axs[0].plot(a[0], a[1][:, 0])
axs[0].set_xticks([])
axs[0].set_ylabel("Membrane potential (mV)")
axs[1].plot(a[0], I_ext)
axs[1].set_xlabel("Time (ms)")
axs[1].set_ylabel("I_ext ($uA/cm^2$)")
plt.tight_layout()
plt.show()

unstable_limit_cycle = model.find_unstable_limit_cycle(4)
plt.figure()
plt.plot(a[1][:, 0], a[1][:, 1], alpha = 0.9)
plt.plot(unstable_limit_cycle[0], unstable_limit_cycle[1], color='red')
plt.title("Phase plot")     
plt.xlabel("Membrane potential (mV)")   
plt.ylabel("n") 
plt.show()  
```
